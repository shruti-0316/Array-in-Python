**1.Addition of two matris take input from user side.**

---



```python
matrix1 = [
    [1, 2, 3],
    [4, 5, 6]
]

matrix2 = [
    [2, 3, 4],
    [5, 6, 7]
]

result = [
    [0, 0, 0],
    [0, 0, 0]
]

for i in range(len(result)):
 
    for j in range(len(result[i])):
        result[i][j] = matrix1[i][j]+matrix2[i][j]

print(result)
```

**2.Python Program to Transpose a Matrix.**

---



```python
rows = int(input("Enter the Number of rows : " ))
column = int(input("Enter the Number of Columns: "))

print("Enter the elements of Matrix:")
matrix= [[int(input()) for i in range(column)] for i in range(rows)]
print("-------Your  Matrix is---------")

for n in matrix:
    print(n)
result =[[0 for i in range(rows)] for j in range(column)]

for r in range(rows):
   for c in range(column):

       result[c][r] = matrix[r][c]

print("Transpose matrix is: ")
for r in result:
    print(r)
```

**3.Write a python program to sort an array elements using bubble sort.**

---



```python
import array
def bubblesort(array):
  for i in range(len(array)):
    for j in range(len(array)-i-1):
        if array[j] > array[j+1]:
            temp=array[j]
            array[j]=array[j+1]
            array[j+1]=temp

elements=[5,6,4,3,2,1]

print("before sorting array elements :")
for ele in elements:
  print(ele)

bubblesort(elements)

print("After sorting array elements : ")
for ele in elements:
  print(ele)
```

**5.Python Program to Count Even and Odd Numbers in an Array.**

```python
import numpy as np

evenOddArr = np.array([10, 99, 22, 50, 77, 22, 112, 19])
print("The List of Numbers in evenOddArr Array = ", evenOddArr)

evenArrCount = 0
oddArrCount = 0

for i in range(len(evenOddArr)):
    if (evenOddArr[i] % 2 == 0):
        evenArrCount = evenArrCount + 1
    else:
        oddArrCount = oddArrCount + 1

print("The Count of Even Numbers in evenOddArr Array = ", evenArrCount)
print("The Count of Odd Numbers in evenOddArr Array  = ", oddArrCount)
```
